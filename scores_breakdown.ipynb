```python
y_correct = np.load("correct_on_test_set_unscaled.npy")
print(len(y_correct))
```

```python
# human_low = []
human_low = []
human_high = []

ai_low = []
ai_high = []

for i in range(len(test_df)):
    human_result = y_correct[2*i]
    ai_result = y_correct[2*i + 1]
    quality_score = test_df.iloc[i]["Overall"]
        
    if 3.0 <= quality_score <= 7.0:
        human_low.append(human_result)
        ai_low.append(ai_result)
        
    elif 7.0 < quality_score <= 9.0:
        human_high.append(human_result)
        ai_high.append(ai_result)
        
    else:
        assert False
```

##### print("Accuracy on Low Human Essays", np.mean(human_low))
print("Accuracy on High Human Essays", np.mean(human_high))
