<h3>Exploratory Data Analysis</h3>

```python
# load dataset
human_df = pd.read_csv("ielts_writing_dataset.csv")
```

```python
# show length of the dataset
len(human_df)
```

```python
# show columns
for col in human_df:
    print(col, human_df[col].dtype)
    print("------------")
```

```python
# show unique values of some columns
print("Task_Type Values:")
print(human_df["Task_Type"].unique())
print("----------------")
```

```python
print("Question Values:")
for i, question in enumerate(human_df["Question"].unique()):
    print(f"Prompt {i+1}")
    print(question)
    print("----------------")
```

```python
# count number of responses per question
count_df = (
    human_df.groupby('Question')['Essay']
    .nunique()
    .sort_values(ascending=False)
    .reset_index()
    .rename(columns={'Essay': '# Unique Responses'})
)
print(count_df)
```

```python
plt.hist(count_df["# Unique Responses"])
```

```python
# examine some essay responses
for response in human_df["Essay"].head():
    print(response)
    print("------------")
```

<h3>Cleaning</h3>

```python
# drop all rows with task 1, which correspond to essays that reference external graphs not available to me
mask = human_df["Task_Type"] == 2
human_df = human_df[mask]
print(len(human_df))
```

```python
# reprint unique questions to check for outliers
print("Question Values:")
for i, question in enumerate(human_df["Question"].unique()):
    print(f"Prompt {i+1}")
    print(question)
    print("----------------")
```
